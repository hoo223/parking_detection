```python
import os
```

```python
folders = ['fold1', 'fold2', 'fold3', 'fold4', 'fold5']
for folder in folders:
    entries = os.listdir('./'+folder)
    print(folder, entries)
    
    empty_list = open(folder+'_empty.txt', 'w')
    occupied_list = open(folder+'_occupied.txt', 'w')
    all_list = open(folder+'_all.txt', 'w')

    for entry in entries:
        print(entry)
        if 'checkpoints' not in entry:
            lines = open('./'+folder+'/'+entry, 'r').readlines()
            if 'all' in entry:
                all_list.writelines(lines)
            elif 'empty' in entry:
                empty_list.writelines(lines)
            elif 'occupied' in entry:
                occupied_list.writelines(lines)

    empty_list.close()
    occupied_list.close()
    all_list.close()
```
