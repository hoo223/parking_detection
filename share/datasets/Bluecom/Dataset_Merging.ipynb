```python
import os 

path = './total_20220724_20220818_update'
annotation_list = os.listdir(path) 
annotation_list.remove('images')
print(annotation_list)
```

```python
import xml.etree.ElementTree as ET

dataset_numbering = {}
for annotation in annotation_list:
    # annotation 파일 불러오기
    filePath = os.path.join(path, annotation)
    tree = ET.parse(filePath)
    root = tree.getroot()
    dataset_name = root.find("meta").find("task").find("name").text
    dataset_name = dataset_name.replace(' ','')
    dataset_numbering[dataset_name] = annotation.split('.')[0].split('s')[1]
    print(annotation, dataset_name)
```

### 각 폴더에 있는 데이터셋 모으기

```python
import shutil
```

```python
root = './data_export_20221028'
dataset_folder = []
for item in os.listdir(root):
    if 'zip' not in item:
        print(item)
        dataset_folder.append(os.path.join(root, item))
```

```python
dataset_folder
```

```python
dst_image = os.path.join(root, 'images')
dst_xml = root

for i, folder in enumerate(dataset_folder):
    # Numbering
    
    
    # Merging
    src = os.path.join(folder, 'images')
    for item in os.listdir(src):
        #print(src, dst)abs
        try:
            shutil.move(os.path.join(src, item), dst_image)
        except:
            pass
    
    try:
        os.rename(os.path.join(folder, 'annotations.xml'), os.path.join(folder, 'annotations{}.xml'.format(i+1)))
    except:
        pass
    src = os.path.join(folder, 'annotations{}.xml'.format(i+1))
    print(src, dst_xml)
    try:
        shutil.move(src, dst_xml)
    except:
        pass
```
