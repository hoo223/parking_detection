### all, empty, occupied overlap removed txt (with 5-fold)

```python
import random

label_file_total_20221028_all = open(os.path.join('total_20221028_generated', 'total_20221028_all.txt'), 'r')
label_file_total_20221028_empty = open(os.path.join('total_20221028_generated', 'total_20221028_empty.txt'), 'r')
label_file_total_20221028_occupied = open(os.path.join('total_20221028_generated', 'total_20221028_occupied.txt'), 'r')
label_file_empty_all = open(os.path.join('empty_generated', 'empty_all.txt'), 'r')


label_file_valid = open(os.path.join('total_20221028+empty_generated', 'total_20221028+empty_generated_overlap_removed_valid.txt'), 'w')
label_file_test = open(os.path.join('total_20221028+empty_generated', 'total_20221028+empty_generated_overlap_removed_test.txt'), 'w')



blacklist = ['시티필드_2',
             '송도국제업무지구+c8-2블럭+업무복합시설',
             '대륭테크노타운8차',
             '커널워크+겨울동_3',
             '청학동하나아파트',
             '어바니엘가산',
             '현대프리미엄아울렛송도점',
             '연수두손이리옴프라자ⅱ_1',
             '센트럴파크_1',
             '송도복합상가_2',
             '서광빌라트',
             '뉴가산시티',
             '시티필드_3',
             '이토타워',
             'New 늘봄 공원 지하 공영주차장',
             '트리플스트리트'
            ]

empty_list = []
occupied_list = []

# empty
total_20221028_empty_lines = label_file_total_20221028_empty.readlines()
empty_all_lines = label_file_empty_all.readlines()

for line in total_20221028_empty_lines:
    remove = False
    for task in blacklist:
        if task in line:
            #print('here')
            remove = True
    if not remove:
        print(line)
        empty_list.append(line)
        
for line in empty_all_lines:
    remove = False
    for task in blacklist:
        if task in line:
            #print('here')
            remove = True
    if not remove:
        print(line)
        empty_list.append(line)

# occupied
total_20221028_occupied_lines = label_file_total_20221028_occupied.readlines()

for line in total_20221028_occupied_lines:
    remove = False
    for task in blacklist:
        if task in line:
            #print('here')
            remove = True
    if not remove:
        print(line)
        occupied_list.append(line)
```

```python
len(empty_list), len(occupied_list)
```

```python
# shuffling
random.shuffle(empty_list)
random.shuffle(occupied_list)
```

```python
# save
valid_size = round(len(empty_list)*0.3)
print(len(empty_list[:valid_size]), len(empty_list[valid_size:]))
label_file_valid.writelines(empty_list[:valid_size])
label_file_test.writelines(empty_list[valid_size:])

valid_size = round(len(occupied_list)*0.3)
print(len(occupied_list[:valid_size]), len(occupied_list[valid_size:]))
label_file_valid.writelines(occupied_list[:valid_size])
label_file_test.writelines(occupied_list[valid_size:])
```
